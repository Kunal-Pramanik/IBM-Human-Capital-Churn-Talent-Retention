# Exploratory Data Analysis

```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("IBM_HR_Data_Enterprise.csv")
```

```
df.describe()
```

```
df["Attrition"].value_counts().plot(kind="bar")
plt.show()
```

```
df.groupby("Department")["AttritionFlag"].mean().plot(kind="bar")
plt.show()
```
